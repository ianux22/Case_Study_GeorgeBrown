## PCA

### Standardization

Let's standardize the variables

```python
# exclude a data type
df_num = tayko_df[['Freq','last_update_days_ago','1st_update_days_ago','Spending']]
df_cat = tayko_df.drop(['Freq','last_update_days_ago','1st_update_days_ago','Spending'], axis=1)

import pandas as pd
from sklearn import preprocessing

x = df_num.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_scaled = pd.DataFrame(x_scaled)
df_scaled.columns = ['Freq', 'last_update_days_ago', '1st_update_days_ago', 'Spending']


scaled_df = pd.concat([df_scaled.reset_index(drop=True),df_cat.reset_index(drop=True)], axis=1)

scaled_df
```

### PCA Analysis and evaluation

```python
from sklearn.decomposition import PCA
import numpy as np
pca = PCA()
pca.fit(scaled_df)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1

print('Number of PC necessary to have 95% of explained variance: ', d)
```

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

plt.figure(figsize=(9,9))
plt.plot(cumsum, linewidth=3)
plt.axis([0, 30, 0, 1])
plt.xlabel("Dimensions")
plt.ylabel("Explained Variance")
plt.plot([d, d], [0, 0.95], "k:")
plt.plot([0, d], [0.95, 0.95], "k:")
plt.plot(d, 0.95, "ko")
plt.annotate("Elbow", xy=(65, 0.85), xytext=(70, 0.7),
             arrowprops=dict(arrowstyle="->"), fontsize=16)
plt.grid(True)
plt.show()
```

```python
print('cumulated count of the variance explained by PC')
pd.DataFrame(np.array([pca.explained_variance_ratio_[:i].sum() for i in range(1, d+1)]).round(3))
```

9 PC can explain 82.7% of the variance. It seems a good trade-off. Let's use 9 PC

Here is how the dataset appear once the PCA are applied

```python
pca = PCA(n_components=9)
principalComponents = pca.fit_transform(scaled_df)
tayko_PCA = pd.DataFrame(principalComponents, columns=['PC1', 'PC2', 'PC3','PC4', 'PC5',
                                                       'PC6', 'PC7','PC8', 'PC9'])

tayko_PCA.index=scaled_df.index

tayko_PCA
```

Instead here we have the loading matrix

```python
loadings = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2', 'PC3','PC4', 'PC5',
                                                       'PC6', 'PC7','PC8', 'PC9'],
             index = scaled_df.columns.values.tolist())
print('Loading matrix')
loadings
```

### Correlation between loadings and variables

```python
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

loading_matrix = pd.DataFrame(loadings, columns=['PC1', 'PC2', 'PC3','PC4', 'PC5',
                                                       'PC6', 'PC7','PC8', 'PC9'],
             index = scaled_df.columns.values.tolist())
loading_matrix
```

```python
# get correlation matrix plot for loadings
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = pyplot.subplots(figsize=(25,15))

ax = sns.heatmap(loading_matrix, annot=True, cmap='Spectral')
plt.show()
```

## Variable selection via REcursive feature eliminaion (RFE)

Let's try to identify the 5 best variables to predict Purchase

To do so we are going to use the Recursive Feature Elimination (RFE). RFE works by searching for a subset of features by starting with all features in the training dataset and successfully removing features until the desired number remains. We'll try to find the best 5 variables

https://machinelearningmastery.com/rfe-feature-selection-in-python/

```python
# test classification dataset
from sklearn.datasets import make_classification
```

```python
# evaluate RFE for classification
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
# create pipeline
rfe = RFE(estimator=DecisionTreeClassifier(), n_features_to_select=5)
model = DecisionTreeClassifier()
pipeline = Pipeline(steps=[('s',rfe),('m',model)])
# evaluate model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(pipeline, X_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
# fit RFE
rfe.fit(X_train, y_train)
# summarize all features
for i in range(X_train.shape[1]):
    print('Column: %d, Selected %s, Rank: %.3f' % (i, rfe.support_[i], rfe.ranking_[i]))
```
