## PCA

Let's standardize the variables

```python
# exclude a data type
df_num = tayko_df[['Freq','last_update_days_ago','1st_update_days_ago','Spending']]
df_cat = tayko_df.drop(['Freq','last_update_days_ago','1st_update_days_ago','Spending'], axis=1)

import pandas as pd
from sklearn import preprocessing

x = df_num.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_scaled = pd.DataFrame(x_scaled)
df_scaled.columns = ['Freq', 'last_update_days_ago', '1st_update_days_ago', 'Spending']


scaled_df = pd.concat([df_scaled.reset_index(drop=True),df_cat.reset_index(drop=True)], axis=1)

scaled_df
```

```python
from sklearn.decomposition import PCA
import numpy as np
pca = PCA()
pca.fit(scaled_df)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1

print('Number of PC necessary to have 95% of explained variance: ', d)
```

The graph below shows the cumulated variance as function of the PC.

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

plt.figure(figsize=(9,9))
plt.plot(cumsum, linewidth=3)
plt.axis([0, 30, 0, 1])
plt.xlabel("Dimensions")
plt.ylabel("Explained Variance")
plt.plot([d, d], [0, 0.95], "k:")
plt.plot([0, d], [0.95, 0.95], "k:")
plt.plot(d, 0.95, "ko")
plt.annotate("Elbow", xy=(65, 0.85), xytext=(70, 0.7),
             arrowprops=dict(arrowstyle="->"), fontsize=16)
plt.grid(True)
plt.show()
```

```python
print('cumulated count of the variance explained by PC')
pd.DataFrame(np.array([pca.explained_variance_ratio_[:i].sum() for i in range(1, d+1)]).round(3))
```

```python
pca = PCA(n_components=9)
principalComponents = pca.fit_transform(scaled_df)
tayko_PCA = pd.DataFrame(principalComponents, columns=['PC1', 'PC2', 'PC3','PC4', 'PC5',
                                                       'PC6', 'PC7','PC8', 'PC9'])

tayko_PCA.index=scaled_df.index

tayko_PCA
```

Loading matrix

```python
loadings = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2', 'PC3','PC4', 'PC5',
                                                       'PC6', 'PC7','PC8', 'PC9'],
             index = scaled_df.columns.values.tolist())
print('Eigenvectors')
loadings
```

```python
y_OLS_train = df_train.Spending.copy()
X_OLS_train = df_train.drop(['Spending', 'Purchase'], axis=1)

y_OLS_test = df_test.Spending.copy()
X_OLS_test = df_test.drop(['Spending', 'Purchase'], axis=1)
```

We are going to build an OLS model with variabes chosen via stepwise selection, in order to have the best OLS model.

```python
import statsmodels.api as sm 

def stepwise_selection(data, target,SL_in=0.05,SL_out = 0.05):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<SL_in):
            best_features.append(new_pval.idxmin())
            while(len(best_features)>0):
                best_features_with_constant = sm.add_constant(data[best_features])
                p_values = sm.OLS(target, best_features_with_constant).fit().pvalues[1:]
                max_p_value = p_values.max()
                if(max_p_value >= SL_out):
                    excluded_feature = p_values.idxmax()
                    best_features.remove(excluded_feature)
                else:
                    break 
        else:
            break
    return best_features
```

```python
stepwise_selection(X_OLS_train,y_OLS_train)
```

```python
from statsmodels.formula.api import ols

fit = ols('Spending ~  Freq + + C(Address_is_res) + C(source_h) + C(Web_order) + last_update_days_ago + C(source_c) + C(source_a) + C(source_u) -1', data=df_train).fit()  # we need to specify CHAS as categorical variable

print(fit.summary())
```

The R-squared of the model is around 60%, that is enough for an OLS model. All the p-values are significative (under the threshold of 0.05; C(Address_is_res)[0] has a larger P-value but since C(Purchase)[1] is significant we cannot remove it) so it menas the variables are useful to explain the target variable.

We can see that "Freq" has the highest coefficient, followed by "source_h" and "source_c".

```python
predictions = (fit.predict(X_OLS_test))
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import math
print('Mean squared error: ', mean_squared_error(y_OLS_test, predictions))
print('Root mean squared error: ', math.sqrt(mean_squared_error(y_OLS_test, predictions)))
print('Mean absolute error: ', mean_absolute_error(y_OLS_test, predictions))
```
